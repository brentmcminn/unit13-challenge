```python
import pandas as pd 
import seaborn as sns 
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
import statsmodels.formula.api as smf
```

```python
#left is respone variable Right = predictor 

model = smf.ols(formula='tip ~ total_bill + size', data=tips)
```

```python
results = model.fit()

print(results.summary())
```

```python


results.params
```

```python
#confience interval identifies the posible values the estimated value can take on

#values less than [.025 .975]
results.conf_int()
```

```python
from sklearn import linear_model

lr = linear_model.LinearRegression()
```

```python
#error bc it wasnts 2D array and it only received 1D arrray
#we can see the error because [] indicated 1D array -- it needs [[]]

predicted = lr.fit(X=tips['total_bill'].values.reshape(-1,1),y=tips['tip'])
```

```python
print(predicted.coef_)
```

```python
predicted.intercept_
```
